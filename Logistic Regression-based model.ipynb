```python
reload(pu)
s = pu.season(json.load(open('2016_college_men.json','r')))
```

```python
s.games_to_LR()
```

```python
clf = LogisticRegression(C=1.0, penalty='l2', tol=0.01)
clf.fit(s.LR['X'],s.LR['y'])
```

```python
delta_ec = [[dec] for dec in np.linspace(-10,10)]
p = clf.predict_proba(delta_ec)
```

```python
%matplotlib inline
```

```python
plt.plot([dec[0] for dec in delta_ec],p[:,1],'ro-')
plt.xlabel('$log_{10}(EC_1) - log_{10}(EC_2))$')
plt.ylabel('$P($Team 1 wins$)$')
plt.title('Logistic Regression on Eigenvector Centrality')
plt.show()
```
